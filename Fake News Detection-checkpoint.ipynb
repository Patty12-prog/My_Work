```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import re
import string
```

```python
df_fake = pd.read_csv("Fake.csv")
df_true = pd.read_csv("True.csv")
```

```python
df_fake.head(10)
```

```python
df_true.head(10)
```

```python
df_fake["class"] = 0
df_true["class"] = 1
```

```python
df_fake.shape,df_true.shape
```

```python
df_fake_manual_testing = df_fake.tail(10)
for i in range(23480,23470,-1):
    df_fake.drop([i],axis = 0,inplace = True)
df_true_manual_testing = df_true.tail(10)
for i in range(21416,21406,-1):
    df_true.drop([i],axis = 0,inplace = True)
```

```python
df_manual_testing = pd.concat([df_fake_manual_testing, df_true_manual_testing], axis=0)
df_manual_testing.to_csv("manual_testing.csv")
```

```python
df_merge = pd.concat([df_fake,df_true], axis = 0)
df_merge.head(10)
```

```python
df = df_merge.drop(["title","subject","date"],axis = 1)
df.head(10)
```

```python
df = df.sample(frac = 1)
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
def word_drop(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"\[.*?\]", " ", text)  # Remove text inside square brackets
    text = re.sub(r"https?://\S+|www\.\S+", " ", text)  # Remove URLs
    text = re.sub(r"<.*?>", " ", text)  # Remove HTML tags
    text = re.sub(r"[%s]" % re.escape(string.punctuation), " ", text)  # Remove punctuation
    text = re.sub(r"\n", " ", text)  # Remove newlines
    text = re.sub(r"\w*\d\w*", " ", text)  # Remove words containing numbers
    text = re.sub(r"\s+", " ", text).strip()  # Remove extra spaces
    return text
```

```python
df["text"] = df["text"].apply(word_drop)
```

```python
df.head(10)
```

```python
x = df["text"]
y = df["class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
LR.score(xv_test, y_test)
```

```python
pred_LR = LR.predict(xv_test)
```

```python
print(classification_report(y_test, pred_LR))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
DT.score(xv_test, y_test)
```

```python
pred_DT = DT.predict(xv_test)
```

```python
print(classification_report(y_test, pred_DT))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
GBC = GradientBoostingClassifier(random_state = 0)
GBC.fit(xv_train, y_train)
```

```python
GBC.score(xv_test, y_test)
```

```python
pred_GBC = GBC.predict(xv_test)
```

```python
print(classification_report(y_test, pred_GBC))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC = RandomForestClassifier(random_state = 0)
RFC.fit(xv_train, y_train)
```

```python
RFC.score(xv_test, y_test)
```

```python
pred_RFC = RFC.predict(xv_test)
```

```python
print(classification_report(y_test,pred_RFC))
```

```python
def output_label(n):
    if n == 0:
       return "Fake News"
    elif n == 1:
       return "True News"
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(word_drop)
    new_x_test=new_def_test["text"]
    new_xv_test=vectorization.transform(new_x_test)
    pred_LR=LR.predict(new_xv_test)
    pred_DT=DT.predict(new_xv_test)
    pred_GBC=GBC.predict(new_xv_test)
    pred_RFC=RFC.predict(new_xv_test)
    print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_label(pred_LR),output_label(pred_DT),output_label(pred_GBC),output_label(pred_RFC)))
```
